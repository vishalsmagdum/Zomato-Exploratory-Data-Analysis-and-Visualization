```python
# As we can see 'Country Code' column is common in both datasets so we can merge 2 datasets.
# on='' will ask on which column we need to join 2 datasets.
# how='' basically does joining like left, right inner.
final_df = pd.merge(df,df_country, on = 'Country Code', how='left') # here we need to focus more on left table.
```

```python
# we combined 2 datasets based on column commons and saved as new data as 'final_df'.

final_df.head()
```

```python
# To check Data Types

final_df.dtypes
```

```python
final_df.columns
```

```python
# to check specific count of values in paraticular column.

final_df.Country.value_counts()

# We can observe that Zomato is operated more in India, so large number of transactions will be in India than other countries.
```

```python
# to check values in specific column
# we are going to save it in a variable 'country_names'
country_names=final_df.Country.value_counts().index
```

```python
country_names
```

```python
# count fo values in that values according to each country

country_values=final_df.Country.value_counts().values
country_values
```

```python
# We basically did this to plot a Pie chart

plt.pie(country_values, labels = country_names)
plt.show()
```

```python
# Which are top 3 countries that uses Zomato with percentage.
# autopct- if we write any number instead of 2 decimal values will get changed.
plt.pie(country_values[:3], labels = country_names[:3],autopct='%1.2f%%')
plt.show()
```

Oservations: Zomato maximum records or transactions are from India. 
After that USA and then United Kingdom.

```python
final_df.columns
```

```python
# Basically this groupby function will show relationship among particular column.
# like rating from 1.8 to 2.4 can be consider as poor rating.
# white is 0 rating with 2148 records.
# we renamed last column also which is count.
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

## Observations:
1. When Rating is between 4.5 to 4.9 ---> Excellent
2. When Rating is between 4.0 to 4.4 ---> Very Good
3. Whe Rating is between 3.5 to 3.9 ---> Good
4. When Rating is between 2.5 to 3.4 ---> Average
5. When Rating is between 2.0 to 2.4 ---> Poor

```python
ratings.head()
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize'] = (24,18)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize'] = (24,18)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette = ['Blue','red','orange','Yellow','Green','pink'])
```

Observations:
    1. Not Rated count is very high.
    2. Maximum number of rating are between  2.5  to 3.4


```python
## Count Plot
sns.countplot(x = 'Rating color', data = ratings, palette = ['Blue','red','orange','Yellow','Green','pink'])
```
